```python
#Clean Crop Recommendation Dataset
import pandas as pd

# Load raw dataset
crop_df = pd.read_csv("../data/raw/Crop_recommendation.csv")

# Check crops
print("Total crops:", crop_df["label"].nunique())

# Remove duplicates (if any)
crop_df = crop_df.drop_duplicates()

# Ensure correct column names
crop_df = crop_df[[
    "N", "P", "K",
    "temperature",
    "humidity",
    "ph",
    "rainfall",
    "label"
]]

# Save clean dataset
crop_df.to_csv("../data/processed/crop_clean.csv", index=False)

print("crop_clean.csv generated successfully")
```

```python
#Clean Disease Dataset

disease = pd.read_csv("../data/raw/plant_disease_dataset.csv")

disease["risk"] = disease["disease_present"].map({0: "Low", 1: "High"})

disease = disease[["temperature","humidity","rainfall","soil_pH","risk"]]

disease.to_csv("../data/processed/disease_clean.csv", index=False)

print("Disease dataset cleaned")
disease.head()
```
