```python
import pandas as pd
import numpy as np

np.random.seed(42)

crop_df = pd.read_csv("../data/processed/crop_clean.csv")

data = crop_df[["label","temperature","humidity","rainfall"]].copy()
data.rename(columns={"label": "crop"}, inplace=True)

data["soil_condition"] = np.random.choice(["Dry","Moist","Wet"], len(data))

# Water requirement classification
water_need = {
    "rice": 3, "banana": 3, "coconut": 3, "sugarcane": 3,
    "maize": 2, "cotton": 2, "wheat": 2, "pigeonpeas": 2,
    "mungbean": 2, "blackgram": 2, "lentil": 2, "chickpea": 2, "coffee": 2, "jute": 2,
    "mothbeans": 1, "kidneybeans": 1, "grapes": 1, "pomegranate": 1,
    "orange": 1, "papaya": 1, "muskmelon": 1, "watermelon": 1,
    "apple": 1, "mango": 1
}

def irrigation_rule(row):

    score = water_need.get(row["crop"], 2)

    if row["temperature"] > 34:
        score += 2
    elif row["temperature"] > 30:
        score += 1

    if row["humidity"] > 85:
        score -= 2
    elif row["humidity"] > 70:
        score -= 1

    if row["rainfall"] > 150:
        score -= 2
    elif row["rainfall"] > 60:
        score -= 1

    if row["soil_condition"] == "Dry":
        score += 2
    elif row["soil_condition"] == "Wet":
        score -= 2

    if score <= 1:
        return "3-4 days"
    elif score == 2:
        return "2-3 days"
    elif score == 3:
        return "1-2 days"
    else:
        return "Immediate"

data["irrigation"] = data.apply(irrigation_rule, axis=1)

data.to_csv("../data/processed/irrigation_clean.csv", index=False)

print("irrigation_clean.csv generated successfully")
```
